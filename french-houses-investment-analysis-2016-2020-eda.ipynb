# We filter the data by 'selling only' deleting ((93000 - 88400)/93000) = 5% of the data 

```python
df_c = df[df['nature_mutation'] == 'Vente'] 
df_c
```

# Let's filter the feature 'numero_disposition'

```python
dff_1 = df_c[df_c['numero_disposition'] == 1] 
dff_2 = df_c[df_c['numero_disposition'] == 2] 
dff_3 = df_c[df_c['numero_disposition'] == 3] 
```

#  We want to create a binary classification of logistic local which is generally buy as one local of variables sizes as oppositions to  non logistics commercial which is in manny cases associates with an appartment and a parking (90% of cases in the city) 

# We can consider less than 2500/93000 = 0.026 (2.6%) of the commercials sells are of 3 rooms so we can merged the 2 with 3 numbers of depositions

```python
#Let's merged dff_2 and dff_3 into a merged dff_4
dff_4 = pd.concat([dff_2,dff_3], axis = 0)
dff_4 
```

# We change to  logistic transaction replacing 2 or 3 in 'numero_disposition' by '0'

```python
dff_4.loc[:, 'numero_disposition'] = 0
```

```python
dff_4
```

# #Let's now merged dff_1 and dff_4 into our dataset 

```python

df_final = pd.concat([dff_1,dff_4], axis = 0)
df_final
```

# From this part we deduce that the right way to separate logistic from non logistic commercial is the label 'numero_disposition' that we code to simplify as non logistic '1' and logistic '0'. With the clear goal to predict the non logistic selling, let's visualize the label:

```python
# A clear histogram representation of the numbers of commercials transactions, 4.91% logistic  againt 95.09% non logistic.
total = len(df_final)
plt.figure(figsize=(10,6))

g = sns.countplot(x='numero_disposition', data=df_final, palette='coolwarm')
g.set_title("Label DISTRIBUTION", fontsize = 20)
g.set_xlabel("Label Values", fontsize = 15)
g.set_ylabel("Count", fontsize = 15)
sizes=[] # Get highest values in y
for p in g.patches:
    height = p.get_height()
    sizes.append(height)
    g.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center", fontsize=14) 
g.set_ylim(0, max(sizes) * 1.15) # set y limit based on highest heights

plt.show()
```

# Categorical encoding

```python
df_final.info()
```

# Label encoding is one of many ways to encoded categorical data, for more informations please read my article in toward data science : https://towardsdatascience.com/5-categorical-encoding-tricks-you-need-to-know-today-as-a-data-scientist-73cf75595298

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df_final['type_local'] = labelencoder.fit_transform(df_final['type_local'])
df_final['nom_commune'] = labelencoder.fit_transform(df_final['nom_commune'])
df_final['code_nature_culture'] = labelencoder.fit_transform(df_final['code_nature_culture'])
df_final['nature_culture'] = labelencoder.fit_transform(df_final['nature_culture'])
df_final['nature_culture_speciale'] = labelencoder.fit_transform(df_final['nature_culture_speciale'])
df_final['ancien_id_parcelle'] = labelencoder.fit_transform(df_final['ancien_id_parcelle'])
df_final['ancien_nom_commune'] = labelencoder.fit_transform(df_final['ancien_nom_commune'])
df_final['adresse_suffixe'] = labelencoder.fit_transform(df_final['adresse_suffixe'])
df_final['adresse_code_voie'] = labelencoder.fit_transform(df_final['adresse_code_voie'])
df_final['code_commune'] = labelencoder.fit_transform(df_final['code_commune'])

df_final['adresse_nom_voie'] = labelencoder.fit_transform(df_final['adresse_nom_voie'])
df_final['code_departement'] = labelencoder.fit_transform(df_final['code_departement'])
df_final['code_nature_culture_speciale'] = labelencoder.fit_transform(df_final['code_nature_culture_speciale'])
```

# Features selections 

# Create X variables and y the label we want to predict 

```python
X = dff02.drop(['numero_disposition', 'Unnamed: 0'], axis = 1) 
y = dff02['numero_disposition']
X.shape,y.shape
```

# Use inbuilt class feature_importances of tree based classifiers

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) 
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(11).plot(kind='barh')
plt.show()
```
